```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Question 1
df1=pd.read_csv('day.csv')
df2=pd.read_csv('hour.csv')
```

### There are no missing values

```python
#Question 3
t=0
for i in df1.columns:
    if(i in ['instant','dteday','casual','registered','cnt']):
        continue
    plt.figure(t)
    sns.distplot(df1[i])
    t=t+1
    plt.title('count plot of '+i)
```

```python
#Question 4
t=0
for i in df1.columns:
    if(i in ['instant','dteday','temp','atemp','hum','windspeed','casual','registered','cnt']):
        continue
    plt.figure(t)
    sns.boxplot(x=df1[i],y=df1['cnt'])
    t=t+1
```
