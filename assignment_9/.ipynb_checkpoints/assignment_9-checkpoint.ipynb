```python
import pandas as pd
import numpy as np
```

```python
#Question 1
df1=pd.read_csv("train.csv")
df2=pd.read_csv("test.csv")
df=pd.concat([df1,df2],axis=0,ignore_index=True)
df.head()
```

```python
df.info()
```

```python
#Question 2
df=df[df["Survived"].isnull()==False]
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#Question 3
df["Has_Cabin"]=df["Cabin"].notnull()
```

```python
df.tail()
```

```python
#Question 4
l=[]
for i in df["Name"]:
    s=i.split(',')
    d=str(s[1]).split(' ')
    if(d[1] in ['Mr.','Miss.','Mrs.','Master.']):
        l.append(d[1])
    else:
        l.append('Special')
    print(l)
df["Title"]=l
df.head()
```

```python
df.isnull().sum()
```

```python
#Queston 5
for i in df.columns:
    if(df[i].dtypes==object):
        if i in 'Cabin':
            df[i].fillna(value="C00",inplace=True)
        else:
            df[i].fillna(method="ffill",inplace=True)
    else:
        df[i].fillna(df[i].mean(),inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
#Question 6
df["Categorical_age"]=pd.qcut(df["Age"],4,labels=False)
df["Categorical_fare"]=pd.cut(df["Fare"],4,labels=False)
```

```python
#Question 7
df=df.drop(columns=["Cabin","Name","PassengerId","Ticket","Age","Fare","SibSp","Parch"])
df.head()
```

```python
#Question 8
df=pd.get_dummies(df)
df.columns
```

```python
#Question 9
y=df.iloc[:,1:2]
x=df[['Pclass', 'Has_Cabin', 'Categorical_age','Categorical_fare', 'Embarked_C', 'Embarked_Q', 'Embarked_S',
     'Sex_female', 'Sex_male', 'Title_Master.', 'Title_Miss.', 'Title_Mr.',
       'Title_Mrs.', 'Title_Special']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Question 11
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
#Question 12
print(reg.score(x_test,y_test))
```
