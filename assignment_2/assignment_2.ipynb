```python
import numpy as np
```

```python
#Question 1
t=np.random.rand(10,1)
print("Mean is ",np.mean(t))
```

```python
#Question 2
t=np.random.rand(20,1)
print("Standard deviation is ",np.std(t))
print("Variance is ",np.var(t))
```

```python
#Question 3
A=np.random.rand(10,20)
B=np.random.rand(20,25)
C=np.matmul(A,B)
print(C)
print("Shape is ",np.shape(C))
print("Sum of elements of new matrix is ",C.sum())

```
